```python
# Arithmetic Operations Program

def get_numbers():
    """Get two numbers from the user"""
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    return num1, num2

def perform_operations(num1, num2):
    """Perform arithmetic operations"""
    addition = num1 + num2
    subtraction = num1 - num2
    multiplication = num1 * num2
    division = num1 / num2 if num2 != 0 else "Error: Division by zero!"
    return addition, subtraction, multiplication, division

def main():
    """Main program"""
    num1, num2 = get_numbers()
    addition, subtraction, multiplication, division = perform_operations(num1, num2)
    
    print("Results:")
    print(f"Addition: {addition}")
    print(f"Subtraction: {subtraction}")
    print(f"Multiplication: {multiplication}")
    print(f"Division: {division}")

if __name__ == "__main__":
    main()
```

```python
def analyze_file(filename):
    try:
        with open(filename, 'r') as file:
            text = file.read()
            
            # Count number of words
            num_words = len(text.split())
            
            # Count number of lines
            num_lines = text.count('\n') + 1
            
            # Occurrence of particular word
            word = input("Enter a word to search: ")
            word_count = text.lower().count(word.lower())
            
            # Occurrence of particular character
            char = input("Enter a character to search: ")
            char_count = text.count(char)
            
            # Number of blank spaces
            space_count = text.count(' ')
            
            print(f"Number of words: {num_words}")
            print(f"Number of lines: {num_lines}")
            print(f"Occurrence of '{word}': {word_count}")
            print(f"Occurrence of '{char}': {char_count}")
            print(f"Number of blank spaces: {space_count}")
    
    except FileNotFoundError:
        print("File not found.")

filename = input("Enter the filename: ")
analyze_file(filename)
```

```python
def copy_file():
    # Get file names from user
    src_filename = input("Enter source file name: ")
    dest_filename = input("Enter destination file name: ")

    try:
        # Open source file in read mode and destination file in write mode
        with open(src_filename, 'r') as src_file:
            with open(dest_filename, 'w') as dest_file:
                # Read contents from source file
                contents = src_file.read()
                # Write contents to destination file
                dest_file.write(contents)
        print("File copied successfully!")
    except FileNotFoundError:
        print("Source file not found.")
    except Exception as e:
        print("An error occurred:", str(e))

copy_file()
```

```python
def read_file_in_reverse():
    filename = input("Enter the filename: ")

    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            reversed_lines = lines[::-1]
            
        print("File contents in reverse order:")
        for line in reversed_lines:
            print(line.strip())
    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print("An error occurred:", str(e))

read_file_in_reverse()
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    return merge(left_half, right_half)

def merge(left, right):
    merged = []
    left_index = 0
    right_index = 0
    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1
    merged.extend(left[left_index:])
    merged.extend(right[right_index:])
    return merged

def tim_sort(arr):
    return sorted(arr)  # Using Python's built-in Tim Sort

def main():
    arr = [5, 2, 8, 1, 9]
    print("Original Array:", arr)

    print("Selection Sort:", selection_sort(arr[:]))
    print("Merge Sort:", merge_sort(arr[:]))
    print("Tim Sort:", tim_sort(arr))

if __name__ == "__main__":
    main()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()

    def delete(self, data):
        if self.head is None:
            return

        if self.head.data == data:
            self.head = self.head.next
            return

        current = self.head
        while current.next:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next

# Test the implementation
linked_list = SinglyLinkedList()

# Insert elements
linked_list.insert(1)
linked_list.insert(2)
linked_list.insert(3)
linked_list.insert(4)
linked_list.insert(5)

# Display the linked list
print("Linked List:")
linked_list.display()

# Delete an element
linked_list.delete(3)

# Display the linked list after deletion
print("Linked List after deletion:")
linked_list.display()
```

```python
import threading
import time

def print_numbers():
    for i in range(10):
        time.sleep(1)
        print(i)

def print_letters():
    for letter in 'abcdefghij':
        time.sleep(1)
        print(letter)

# Create threads
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

# Start threads
thread1.start()
thread2.start()

# Wait for threads to finish
thread1.join()
thread2.join()

print("Done!")
```
